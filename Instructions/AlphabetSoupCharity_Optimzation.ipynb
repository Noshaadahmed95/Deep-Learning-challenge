```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("../Resources/charity_data.csv")
application_df.head()
```

```python
application_df.columns
```

```python
# See if the model performs better with just taking EIN column out
application_df = application_df.drop(columns =['EIN'])
application_df.head()
```

```python
application_df.nunique()
```

```python
name_type_counts = application_df['NAME'].value_counts()
name_type_counts
```

```python
# remove values of 1
my_cutoff_range = name_type_counts<100
name_to_replace = list(name_type_counts[my_cutoff_range].index)
print(my_cutoff_range)
```

```python
# Replace in dataframe
for app in name_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
app_classification_counts = application_df['CLASSIFICATION'].value_counts()
app_classification_counts.head(50)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >9
#  YOUR CODE GOES HERE
app_class_counts1 = app_classification_counts>9
app_class_counts1.head(50)
```

```python
my_cutoff_range2 = app_classification_counts<100
classifications_to_replace = list(app_classification_counts[my_cutoff_range2].index)
print(classifications_to_replace)
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df['NAME'].nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
converted_df = pd.get_dummies(application_df)
converted_df.head()
```

```python
# Remove IS_SUCCESSFUL target from the features data
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
X = converted_df.drop(columns='IS_SUCCESSFUL')
y = converted_df['IS_SUCCESSFUL']
```

```python
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
## Compile, Train and Evaluate the Model
```

```python
X.shape
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn_model = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=60, activation='relu', input_dim=88))

# Second hidden layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=60, activation='relu'))

# Output layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Increasing the epochs to 100 from 50
#  YOUR CODE GOES HERE
fir_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
